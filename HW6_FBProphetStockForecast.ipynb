```python
'''
periods期數越長，預測偏離越大，
短期看起來比較貼近實際股價線圖
'''

import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from prophet import Prophet
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt

hist = yf.download("2379.TW", start="2020-01-01", end="2021-07-14")

hist = hist.reset_index()
df = hist[['Date', 'Adj Close']].copy()
df.columns = ['ds', 'y']

plt.figure(figsize=(10, 6))
plt.plot(pd.to_datetime(df['ds']), df['y'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
model = Prophet(changepoint_prior_scale=1)
model.fit(df)
```

```python
# 預測未來 30 天
future = model.make_future_dataframe(periods=10)
forecast = model.predict(future)

df['y'].plot(figsize=(12, 6), color='green', linestyle='-', lw=4, legend='Price')
#forecast['trend'].plot(figsize=(12, 6), color='red', linestyle='-', lw=4, legend='Price')

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False

font = {'family': 'Microsoft JhengHei', 'weight': 'normal', 'size': 20}
fig = model.plot(forecast, xlabel='Date', ylabel=r'Price')
plt.title('預測未來 10 天', fontdict=font)
plt.show()
```
